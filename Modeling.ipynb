```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense, Activation, Embedding, Flatten, LeakyReLU, BatchNormalization, Dropout
from keras.activations import relu, sigmoid
import warnings
warnings.filterwarnings('ignore')
```

Neural networks

```python
def create_model(layers, activation):
    model = Sequential()
    for i, nodes in enumerate(layers):
        if i==0:                                                                
            model.add(Dense(nodes,input_dim=X.shape[1]))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
        else:                                                                    
            model.add(Dense(nodes))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
            
    model.add(Dense(units = 1, kernel_initializer= 'glorot_uniform', activation = 'sigmoid')) 
    
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model
```

```python
ann_model = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
layers = [[16], [32, 16], [64, 32, 16]]
activations = ['sigmoid', 'relu']
param_grid = dict(layers=layers, activation=activations, batch_size = [128, 256], epochs=[30])
grid = GridSearchCV(estimator = ann_model, param_grid = param_grid, scoring = 'f1', cv = 5)
```

```python
grid_result = grid.fit(X, y)
```

```python
print(grid_result.best_score_, grid_result.best_params_)
```

```python
def create_model_actual(layers, activation):
    model = Sequential()
    for i, nodes in enumerate(layers):
        if i==0:                                                                
            model.add(Dense(nodes,input_dim=X[feature_set].shape[1]))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
        else:                                                                    
            model.add(Dense(nodes))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
            
    model.add(Dense(units = 1, kernel_initializer= 'glorot_uniform', activation = 'sigmoid')) 
    
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model
```

```python
ann_model_actual = KerasClassifier(build_fn=create_model_actual, verbose=0)
```

```python
layers = [[16], [32, 16], [64, 32, 16]]
activations = ['sigmoid', 'relu']
param_grid = dict(layers=layers, activation=activations, batch_size = [128, 256], epochs=[30])
grid_actual = GridSearchCV(estimator = ann_model_actual, param_grid = param_grid, scoring = 'f1', cv = 5)
```
